# Modification du data

```python
df = pd.read_csv("Bakery sales.csv")
# Combine 'date' and 'time' columns to create a 'datetime' column
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'])

# Convert 'unit_price' to numeric, removing the '€' symbol and replacing ',' with '.'
df['unit_price'] = df['unit_price'].str.replace('€', '').str.replace(',', '.').astype(float)

# Calculate total price for each row
df['total_price'] = df['Quantity'] * df['unit_price']

# Extract date and time components
df['date'] = df['datetime'].dt.date
df['time'] = df['datetime'].dt.time
```

```python
df.head()
```

# Analyse

```python
# Group by date and calculate daily sales
daily_sales = df.groupby('date')['total_price'].sum().reset_index()

# Plot daily sales
plt.figure(figsize=(12, 6))
plt.plot(daily_sales['date'], daily_sales['total_price'])
plt.title('Daily Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales (€)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by article and sum quantities
top_products = df.groupby('article')['Quantity'].sum().sort_values(ascending=False).head(10)

# Plot top 10 selling products
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar')
plt.title('Top 10 Selling Products')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Extract hour from datetime and group by hour
df['hour'] = df['datetime'].dt.hour
hourly_sales = df.groupby('hour')['total_price'].sum()

# Plot hourly sales pattern
plt.figure(figsize=(10, 6))
hourly_sales.plot(kind='bar')
plt.title('Hourly Sales Pattern')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales (€)')
plt.tight_layout()
plt.show()
```

```python
# Add a week column
df['week'] = df['datetime'].dt.isocalendar().week

# Group by week and calculate weekly sales
weekly_sales = df.groupby('week')['total_price'].sum().reset_index()

# Plot weekly sales trend
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales['week'], weekly_sales['total_price'])
plt.title('Weekly Sales Trend')
plt.xlabel('Week of Year')
plt.ylabel('Total Sales (€)')
plt.tight_layout()
plt.show()
```
